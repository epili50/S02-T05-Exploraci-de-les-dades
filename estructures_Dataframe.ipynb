#### Crear nuevas columnas
Una columna marca la velocidad media de cada vuelo dividiendo la distancia de vuelo en la velocidad y multiplando por 60 para pasar de minutos a horas
La otro columna avisa si los vuelos tuvieron retraso o no y clasifica los mismos en cortos, normales y largos.

```python
#Creo una columan de velocidad mmedia de vuelo millas/h
             
delayFlights["MeanFlight/h"] = delayFlights['Distance']/delayFlights['AirTime']*60
```

```python
delayFlights[['Distance', 'AirTime', 'MeanFlight/h']].head()
```

```python
#Creo columnas que marquen 

conditions = [
    (delayFlights['ArrDelay'] < 15),
    (delayFlights['ArrDelay'] > 15) & (delayFlights['ArrDelay'] < 60),
    (delayFlights['ArrDelay'] > 60) & (delayFlights['ArrDelay'] < 90),
    (delayFlights['ArrDelay'] > 90),
    ]

# create a list of the values we want to assign for each condition
values = ['onTime','shortDelay' ,'delay' , 'longDelay']

# create a new column and use np.select to assign values to it using our lists as arguments
delayFlights['TypesOfDeley'] = np.select(conditions, values, default='Not Specified')
delayFlights[['ArrDelay', 'TypesOfDeley']]
```

#### Ver aerolineas con mayores retrasos

Todas las companías con sus minutos de demoras

```python
#agrupo las companías en companies
companies = delayFlights.groupby('UniqueCarrier')

#imprimo todas las companías con sus acumulados de minutos de retrazo
companies['ArrDelay'].agg(np.sum)

```

Todas las companías con sus minutos de demoras ORDENADAS de mayor a menor

```python
delayCompanies = companies['ArrDelay'].agg(np.sum)

delayCompanies.sort_values(ascending = False)
```

Las companías ordenadas de mayor a menor según el promedio de demoras de sus vuelos

```python
delayCompanies = companies['ArrDelay'].agg(np.mean)

delayCompanies.sort_values(ascending = False)
```

#### Vuelos más largo y más cortos 

```python
#creo un DF sacando las distancias duplicadas
tempDeleyFlights = delayFlights.drop_duplicates(subset = ['Distance'])

#hago que me ordene las distancias de mayor a menor
distanceFlights = tempDeleyFlights.sort_values('Distance', ascending =False)

#Imprimo las diez distancias más largas
distanceFlights[['Dest', 'Origin', 'Distance']].head(10)
```

```python
##Imprimo las diez distancias más cortas

distanceFlights[['Dest', 'Origin', 'Distance']].tail(10)
```
